```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
data=pd.read_csv('/content/fetal_health classification project.csv')
x_train,x_test,y_train,y_test=train_test_split(data.drop('fetal_health',axis=1),data['fetal_health'],test_size=0.2)
```

```python
from lazypredict.Supervised import LazyClassifier
lazy_pred=LazyClassifier(verbose=2,ignore_warnings=True,custom_metric=None)
models,predictions=lazy_pred.fit(x_train,x_test,y_train,y_test)
```
